Define Emotion Labels

```python
# FER2013 emotion labels (7 classes)
emotion_labels = ["Angry", "Disgust", "Fear", "Happy", "Sad", "Surprise", "Neutral"]
```

Image Upload Function

```python
def upload_image():
    print("üìÅ Please upload an image (JPG/PNG):")
    uploaded = files.upload()

    for filename in uploaded.keys():
        print(f"‚úÖ Uploaded: {filename}")
        display(Image(filename))
        return filename
```

Preprocess Image for Model

```python
def preprocess_image(img_path):
    img = cv2.imread(img_path)

    if img is None:
        raise ValueError("‚ùå Error: Unable to load the image!")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_resized = cv2.resize(gray, (48, 48))

    # Normalize & reshape for model
    input_img = face_resized.reshape(1, 48, 48, 1) / 255.0

    return input_img
```

Predict Emotion

```python
def predict_emotion(image):
    predictions = model.predict(image)[0]
    dominant_index = np.argmax(predictions)
    emotion = emotion_labels[dominant_index]
    confidence = predictions[dominant_index] * 100

    print("\nüé≠ Detected Emotion:", emotion)
    print(f"üìä Confidence: {confidence:.2f}%\n")

    # Show probability breakdown
    print("üìà Emotion Breakdown:")
    for i, label in enumerate(emotion_labels):
        print(f"{label:10} ‚Äî {predictions[i]*100:.2f}%")

    return emotion
```

Display Music Recommendations

```python
def show_recommendations(emotion):
    print("\nüéµ Recommended Songs for:", emotion)
    print("------------------------------------------------")

    if emotion in music:
        for song, link in music[emotion]:
            print(f"üé∂ {song}")
            print(f"üîó {link}\n")
    else:
        print("No songs available for this emotion.")
```
